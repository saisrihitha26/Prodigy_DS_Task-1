```python
import requests
import pandas as pd

url = "https://api.worldbank.org/v2/country/all/indicator/SP.POP.TOTL?date=2023&format=json&per_page=300"
response = requests.get(url)
data = response.json()[1]

# Create a DataFrame
df = pd.DataFrame(data)
df = df[['country', 'value']].dropna()
df['country'] = df['country'].apply(lambda x: x['value'])
df.rename(columns={'value': 'Population'}, inplace=True)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.hist(df['Population'], bins=[0, 1e7, 5e7, 1e8, 5e8, 1e9, 1.5e9], edgecolor='black')
plt.title('Distribution of Total Population Across Countries (2023)')
plt.xlabel('Population Size')
plt.ylabel('Number of Countries')
plt.xticks([0, 1e7, 5e7, 1e8, 5e8, 1e9, 1.5e9], ['0-10M', '10-50M', '50-100M', '100-500M', '500M-1B', '1B-1.5B'])
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
